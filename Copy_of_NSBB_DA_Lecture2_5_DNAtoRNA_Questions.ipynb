#### Pythagorean Theorem
Write a function that returns the hypotenouse `c` given side `a` and `b`.

`4 ** 2` returns 16  
`3 ** 3` returns 27

```python
import math as math
def hypothenouse_answer(a, b):

  c = math.sqrt((a ** 2) + (b ** 2))
  return c 
```

```python
hypothenouse_answer(4,4)
```

#### Modulo Operator
Write a function that returns True if the integer is even, and False if the interger is Odd. Hint: use the modulo operator `%` to return the remainder.

`6 / 2` returns `3`; division  
`6 % 2` returns `0`; meaning that the remainder is 0

```python
def even_odd(numerator,denominator):
  if numerator % denominator == 0:
    return "Even"
  return "Odd"
```

```python
even_odd(6,2)
```

#### Count all 4 Nucleotides

```python
def nucleotide_counter(string_entered):
  
  adenine_counter = string_entered.count('A')
  thymine_counter = string_entered.count('T')
  guanine_counter = string_entered.count('G')
  cytosine_counter = string_entered.count('C')

  print ("Count of A's in the entered DNA sequence is : "
                       +  str(adenine_counter))
  print ("Count of T's in the entered DNA sequence is : "
                       +  str(thymine_counter))
  print ("Count of G's in the entered DNA sequence is : "
                       +  str(guanine_counter))
  print ("Count of C's in the entered DNA sequence is : "
                       +  str(cytosine_counter))
```

```python
nucleotide_counter(DNA_string)
```

#### Calculate GC %

```python
def GC_calculator(string_entered):

  adenine_counter = string_entered.count('A')
  thymine_counter = string_entered.count('T')
  guanine_counter = string_entered.count('G')
  cytosine_counter = string_entered.count('C')

  total_counter = adenine_counter + thymine_counter + guanine_counter + cytosine_counter

  g_plus_c = guanine_counter + cytosine_counter

  g_c_calc = guanine_counter / total_counter

  print("GC% is: " + str(g_c_calc,))
```

```python
GC_calculator(DNA_string)
```

#### Convert RNA to AA sequence

```python
def dna_to_aa(entered_string):
       
    codon = entered_string.replace("T","U")
    protein_seq =""
    if len(codon)%3 == 0:
        for i in range(0, len(codon), 3):
            triplet = codon[i:i + 3]
            protein_seq+= AA_dct[triplet]
    return protein_seq
```
